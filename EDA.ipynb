```python
#One hot encoding
#En el label clase 1=venenoso y 0=comestible
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])

onehot_encoder = OneHotEncoder()
features_onehot = onehot_encoder.fit_transform(df.drop('class', axis=1)).toarray()
features_onehot_df = pd.DataFrame(features_onehot, columns=onehot_encoder.get_feature_names_out())

df_final = pd.concat([df['class'], features_onehot_df], axis=1)

df_final.head()
```

```python
import numpy as np
import pandas as pd

# Crear una matriz sin la diagonal superior
corr_unstacked = correlation_matrix.where(
    np.tril(np.ones(correlation_matrix.shape)).astype(bool)
).stack().reset_index()

# Renombrar las columnas para mayor claridad
corr_unstacked.columns = ['Variable 1', 'Variable 2', 'Correlation']

# Filtrar por un umbral si lo deseas
filtered_corr_unstacked = corr_unstacked[
    (corr_unstacked['Correlation'] >= threshold) | 
    (corr_unstacked['Correlation'] <= -threshold)
]

# Mostrar la tabla de correlaciones sin duplicados
print(filtered_corr_unstacked)

#filtered_corr_unstacked.to_csv('corr.csv', index=False)
```
